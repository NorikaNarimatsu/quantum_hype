```python
df_author = pd.read_csv("../data/csv/df_author_combined.csv")

# Count number of unique affiliations per (paperId, researcherId)
affiliation_counts = (
    df_author
    .groupby(['Accession Number (UT)', 'ResearcherID'])['Country']
    .nunique()
    .reset_index(name='num_affiliations')
)

# Sort to find authors with the most affiliations for a single paper
top_multi_affiliations = affiliation_counts.sort_values('num_affiliations', ascending=False)
print("Top authors with the most affiliations on a single paper:")
print(top_multi_affiliations.head(10))
```
